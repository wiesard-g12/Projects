```python
import pandas as pd
df_captions = open('captions.txt', 'r').readlines()
print(df_captions[0:1000])
```

```python
import cv2
df_images=pd.DataFrame()
df_images['image']=[]
for i in range(len(df_captions)):
    path="/Images/"+df_images_captions['image'][i]
    img= cv2.imread(path)
    df_images.loc[i, 'image'] = img
```

```python
# model architecture 
import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(filters=32, kernel_size=(5,5), padding='same', activation='relu', input_shape=(28,28,1)),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    tf.keras.layers.Dropout(0.25),

    tf.keras.layers.Conv2D(filters=64, kernel_size=(3,3), padding='same', activation='relu'),
    tf.keras.layers.Conv2D(filters=64, kernel_size=(3,3), padding='same', activation='relu'),
    tf.keras.layers.MaxPool2D(pool_size=(2,2), strides=(2,2)),
    tf.keras.layers.Dropout(0.25),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dropout(0.25),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer = 'adam', loss='categorical_crossentropy', metrics=["accuracy"])
```

```python
# Create a DataFrame to hold the image-caption pairs
df_images_captions = pd.DataFrame(columns=['image', 'caption'])
for i in range(len(df_captions)):
    df=df_captions[i].split(',')
    
    df_images_captions.loc[i] = [df[0], df[1]]
print(df_images_captions)
```

```python
# caption formatting
for i in range(len(df_images_captions)):

    caption = df_images_captions['caption'][i]
    # Remove leading/trailing spaces, full stops, newlines, and backspaces
    caption = caption.strip().replace('.', '').replace('\n', '').replace('\b', '')
    df_images_captions.at[i, 'caption'] = "[start]" + caption + "[end]"

```
